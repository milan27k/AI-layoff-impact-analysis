```python
###Q1-What is average workforce reduction
df['pct_workforce_cut'].mean()
```

```python
###Q2-Which company has highest layoff
df.groupby('company')['jobs_cut'].sum().sort_values(ascending=False).head(1)
```

```python
###Q3-Which category is most affected
df.groupby('sector')['jobs_cut'].sum().sort_values(ascending=False).head(1)
```

```python
###Q4-What percentage of layoffs is contributed by each company?
total = df['jobs_cut'].sum()
df.groupby('company')['jobs_cut'].sum().apply(lambda x:(x/total)*100).sort_values(ascending=False).head(5)
```

```python
###Q5-Which region is most affected by layoffs
df.groupby('region')['jobs_cut'].sum().sort_values(ascending=False).head(2)
```
